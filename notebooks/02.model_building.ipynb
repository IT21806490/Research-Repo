```python
import os
import glob
import cv2
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
```

```python
# Define paths to the training and validation sets
base_path = 'C:/Users/Admin/Documents/GitHub/Research-Repo/artifacts/dataset_updated/'
training_path = os.path.join(base_path, 'training_set')
validation_path = os.path.join(base_path, 'validation_set')

categories = ['drawings', 'engraving', 'iconography', 'painting', 'sculpture']
```

```python
# Function to load images from a given path
def load_images_from_folder(folder_path):
    images = []
    labels = []
    for category in categories:
        path = os.path.join(folder_path, category).replace('\\', '/')
        print(f"Checking category: {category} in path: {path}")  # Debug print
        if not os.path.exists(path):
            print(f"Path does not exist: {path}")
            continue
        print(f"Contents of {path}: {os.listdir(path)}")  # List directory contents

        for img_path in glob.glob(os.path.join(path, '*.jpg')):
            print(f"Found image: {img_path}")  # Debug print
            img = cv2.imread(img_path)
            if img is not None:
                img = cv2.resize(img, (128, 128))  # Resize images to 128x128
                images.append(img)
                labels.append(category)
    print(f"Loaded {len(images)} images from {folder_path}")
    return images, labels
```

```python
import os

# Define paths using raw strings
training_path = r"C:\Users\Admin\Documents\GitHub\Research-Repo\artifacts\dataset\dataset_updated\training_set"
validation_path = r"C:\Users\Admin\Documents\GitHub\Research-Repo\artifacts\dataset\dataset_updated\validation_set"

categories = ['drawings', 'engraving', 'iconography', 'painting', 'sculpture']

for category in categories:
    train_category_path = os.path.join(training_path, category)
    val_category_path = os.path.join(validation_path, category)
    
    print(f"Contents of {train_category_path}:")
    if os.path.exists(train_category_path):
        print(os.listdir(train_category_path))
    else:
        print(f"Path does not exist: {train_category_path}")
    
    print(f"Contents of {val_category_path}:")
    if os.path.exists(val_category_path):
        print(os.listdir(val_category_path))
    else:
        print(f"Path does not exist: {val_category_path}")
```

```python
# Load training and validation images
train_images, train_labels = load_images_from_folder(training_path.replace('\\', '/'))
val_images, val_labels = load_images_from_folder(validation_path.replace('\\', '/'))
```

```python
# Convert lists to numpy arrays
train_images = np.array(train_images)
val_images = np.array(val_images)   
```

```python
# Encode labels to integers
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)
val_labels_encoded = label_encoder.transform(val_labels)  
```

```python
# Debug print statements to check encoded labels
print(f"Encoded training labels: {train_labels_encoded}")
print(f"Encoded validation labels: {val_labels_encoded}")
```

```python
# Convert labels to one-hot encoding
num_classes = len(categories)
train_labels_one_hot = to_categorical(train_labels_encoded, num_classes=num_classes)
val_labels_one_hot = to_categorical(val_labels_encoded, num_classes=num_classes)
```

```python
# Normalize image data to 0-1 range
train_images = train_images.astype('float32') / 255.0
val_images = val_images.astype('float32') / 255
```

```python
# Debug print statements to check data shapes
print(f"Shape of train_images: {train_images.shape}")
print(f"Shape of train_labels_one_hot: {train_labels_one_hot.shape}")
print(f"Shape of val_images: {val_images.shape}")
print(f"Shape of val_labels_one_hot: {val_labels_one_hot.shape}")
```

```python
# Define the CNN model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(len(categories), activation='softmax')
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model with eager execution enabled for debugging purposes
tf.config.run_functions_eagerly(True)
model.fit(train_images, train_labels_one_hot, epochs=10, validation_data=(val_images, val_labels_one_hot))
```
